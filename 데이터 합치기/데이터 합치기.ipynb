```python
import pandas as pd

english_df1 =pd.DataFrame({
    'name':['dongwook','taeho','jimin'],
    'english_score':[50,89,68]
})

english_df2 = pd.DataFrame({
    'name':['yoonsoo','sowon','haeun'],
    'english_score':[88,91,72]
})
```

```python
english_df1
```

```python
english_df2
```

```python
# 위 둘 데이터를 합쳐보자
```

```python
pd.concat([english_df1,english_df2])
#그러나 인덱스가 원래 데이터 프레임의 인덱스를 따오게된다.
```

```python
#ignore_index를 통해 위 문제점을 해결한다.
english_df=pd.concat([english_df1,english_df2],ignore_index=True)
english_df
```

# join
### concat 함수를 사용하여 위 아래 혹은 옆으로 붙여보았다.
### 이번엔 데이터가 각각 다른 것을 합칠때는 join 연산을 사용하는것이 더 좋다.
### 즉 공통으로 들어있는 데이터 기준으로 합친다. 공통 데이터를 key값이라고 한다.
- inner join
- left outer join: 왼쪽 기준 데이터로 합친다.
- Right outer join
- full outer join : 양 쪽 모든 정보가 들어가며 이번에도 마찬가지로 정보가 없다면 결측값으로 들어간다.

```python
import pandas as pd

employee_df =pd.read_csv('employee.csv')
survey_df =pd.read_csv('survey.csv')
employee_df
```

```python
survey_df
```

# 직원 데이터는 100개인데 설문조사 데이터는 76개이다. 즉 모든 직원이 설문에 응한것은 아니다.
# merge

```python
#고유한 부분은 사원번호이기 떄문에 key값은 id이다.
# 어떤 건 데이터 리스트 묶고 어떤건 안묶는다 이 차이점 기억하기
pd.merge(employee_df,survey_df,on ='id')
```

```python
pd.merge(employee_df,survey_df,on ='id',how='left')
#데이터 100개인거 확인
```

```python
pd.merge(employee_df,survey_df,on ='id',how='right')
#데이터 76개 확인
```

```python
pd.merge(employee_df,survey_df,on ='id',how='left',suffixes=('','_yyy'))
# departement는 병합시 x,y가 붙는 다. 이유는 하나의 프레임에 같은 열 네임이 올수 없기 때문이다.
# 이상태에서 surffixes를 사용해본다.
#surffixes를 사용하여 자동으로 공통적인 데이터 명을 바꿔준다.
# department_x,y가_뒤가 위 내용로 바뀌었다.
```

```python
# key 값이 서로 다른 경우
survey_df=survey_df.rename(columns={'id':'employee_id'})
survey_df
```

```python
company_df=pd.merge(employee_df,survey_df,left_on='id',right_on='employee_id')
#그러나 위와는 다르게 각 열 네임이 살아있다.
company_df
```

# join
- 인덱스를 key값으로 하여 병합한다.

```python
employee_df=employee_df.set_index('id')
survey_df =survey_df.set_index('employee_id')
employee_df.join(survey_df) #앞 데이터는 왼쪽 데이터 뒤는 오른쪽 데이ㅓ이다.
#그러나 실행하면 오류가 난다. 
# 이유는 겹치는 컬럼이 있는데 suffix가 없기 때문이다. 즉 merge에서는 자옹적이지만 
#join에서는 따로 설정해줘야 한다.
```

```python
company_df=employee_df.join(survey_df,lsuffix='_x',rsuffix='_y',how='inner')
#이것도 마찬가지로 둘 중 하나만 설정하는 것도 가능하다.
```

```python
company_df # 두가지 데이터를 조인해서 만든 데이터 프레임이 나온다.
```

```python
# 회사하는 직원만 가져오기
condition=company_df['resign']=='Y'
company_df[condition].describe()
```
