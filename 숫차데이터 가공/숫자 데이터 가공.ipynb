# 스케일: 스케일링을 통해 데이터 규모를 맞춰준다.
## 정규화 : 0, 1로 맞춰준다. (원래값-최솟값)/최대최소차
- 최대값은 1로 바뀌고 최솟값은 0으로 바뀐다.

```python
patient_df['height']=(patient_df['height']-patient_df['height'].min())/(patient_df['height'].max()-patient_df['height'].min())
```

```python
patient_df # height컬럼이 정규화 되었다.
```

```python
patient_df['height'].describe() 
```

# 표준화: 각 값이 평균에 비해 얼마나 크고 작은지 나타낸다.
- 데이터 평균은 0 분산은 1로 변환한다.

```python
(patient_df['height'] - patient_df['height'].mean()) / patient_df['height'].std()
```

```python
patient_df['height']=(patient_df['height'] - patient_df['height'].mean()) / \
patient_df['height'].std()
```

```python
patient_df
```

### 정규화 vs 표준화
- 정규화: 모든 값을 0과1사이로 바꾸어 데이터 범위를 통일하고 싶을때 사용한다.
- 표준화: 평균 기준 떨어진 거리가 중요할 때 사용한다.
