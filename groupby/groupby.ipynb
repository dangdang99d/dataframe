```python
import pandas as pd

netflix_df =pd.read_csv('netflix.csv')
netflix_df
```

```python
netflix_df.info()
#230개 데이터 중 결측값 없음을 확인함
```

```python
netflix_df.describe()
#상영 시간의 평균은 95분 평점 평균은 6.25
#세부적으로 구릅을 나누고 싶다면...
```

```python
netflix_df.groupby('genre')#장르별로 비교함 
#이 뒤에 다른 함수를 사용해줘야 결과물을 인할 수 있다.
```

```python
netflix_df.groupby('genre').count()
```

```python
netflix_df.groupby('genre').size()
```

```python
netflix_df.groupby('genre').min()
# 각 그룹별 각 열의 최솟값이다.
# 문자데이터의 최솟값이다. 최솟값을 구하였고 오름차순으로 정렬 했을때 가장 첫번째로 오게
#title을 보면 알파벳 A로 시작
```

```python
netflix_df.groupby('genre').min(numeric_only=True)
#그럼 숫자 데이터만 보여준다.
```

```python
netflix_df.groupby('genre').mean(numeric_only=True)
```

```python
netflix_df.groupby('genre')['score'].mean()
```

```python
pd.to_datetime(netflix_df['release']).dt.year
#년도 값 추출 그리고 dt쓰는 거 주의사항 없으면 오 
```

```python
netflix_df['year']=pd.to_datetime(netflix_df['release']).dt.year
```

```python
netflix_df.dtypes
```

```python
netflix_df.groupby(['genre','year']).mean(numeric_only=True)
#장르와 year를 인덱싱으로 가지고 있으며 멀티 인덱싱으로 불린다.
```

```python
netflix_df.groupby(['year','genre']).mean(numeric_only=True)
```

```python
netflix_df.groupby(['year','genre']).mean(numeric_only=True).loc[2020,:]
#2020년 정보만 가져오
```

```python
netflix_df.groupby(['year','genre']).mean(numeric_only=True).loc['Drama']
#오류 메세지가 나온다. 이것은 key error이고 인덱싱의 값이 없기 때문이다.
# 데이터 기준 가장 바깥정보만 접근 가능하기 때문이다. 
```

```python
netflix_df.groupby(['year','genre']).mean(numeric_only=True).loc[2020,'Drama']
#이건 가능함
```

```python
# 모든 년도의 드라마 정보 가ㅣㅈ고 오기
years=sorted(netflix_df['year'].unique())
netflix_df.groupby(['year','genre']).mean(numeric_only=True).loc[(years,'Drama'),:]
```

```python
# groupby를 통해 여러가지 통계값을 구하기
#컬럼별로 다른 통계
netflix_df.groupby('genre')['score'].agg(['mean','max','min'])
```

```python
netflix_df.groupby(['genre']).agg({'score':'mean','runtime':['min','max']})
```

 # 피벗테이블

```python
netflix_df.groupby(['year','genre'])['score'].mean()
```

```python
import pandas as pd
pd.pivot_table(netflix_df, values='score',index='year',columns='genre')
# 피벗테이블의 기보본은 평균을 계산해준다.
# groupby는 인덱스에 연도와 장가 들어있었지만 
#피벗 테이블은 인덱스와 컬럼을 모두 사용해준다. index='year',columns='genre'
```

```python
pd.pivot_table(netflix_df, values='score',index='year',columns='genre',aggfunc='max')
```
